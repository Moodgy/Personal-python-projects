```python
def describe_number(number):
    descriptions = {
        "1. Cardinal Number": f"The number itself: {number}",
        "2. Ordinal Number": f"The position of the number in a sequence: {ordinal(number)}",
        "3. Word Form": f"The number written out in words: {number_to_words(number)}",
        "4. Even/Odd": f"Even: {is_even(number)}, Odd: {is_odd(number)}",
        "5. Prime/Composite": f"Prime: {is_prime(number)}, Composite: {is_composite(number)}",
        "6. Positive/Negative/Zero": f"Positive: {is_positive(number)}, Negative: {is_negative(number)}, Zero: {is_zero(number)}",
        "7. Integer/Decimal/Fraction": f"Integer: {is_integer(number)}, Decimal: Not Applicable, Fraction: Not Applicable",
        "8. Rational/Irrational": f"Rational: {is_rational(number)}, Irrational: {is_irrational(number)}",
        "9. Real/Imaginary/Complex": f"Real: {is_real(number)}, Imaginary: {is_imaginary(number)}, Complex: {is_complex(number)}",
        "10. Absolute Value": f"The distance of the number from zero on the number line: {absolute_value(number)}",
        "11. Scientific Notation": f"A method of expressing a very large or very small number: Not Applicable",
        "12. Factors": f"Numbers that can be multiplied together to get the original number: {factors(number)}",
        "13. Multiples": f"Numbers that the original number can be divided by without leaving a remainder: {multiples(number)}",
        "14. Divisibility": f"Whether the number can be divided by another number without leaving a remainder: {divisibility(number)}",
        "15. Square/Cube/Higher Powers": f"The result of multiplying a number by itself: {powers(number)}",
        "16. Roots": f"The inverse operation of raising a number to a given power: {roots(number)}",
        "17. Percentage": f"The number as a fraction of 100: Not Applicable",
        "18. Decimal Places": f"The number of digits after the decimal point: Not Applicable",
        "19. Reciprocal": f"The number flipped upside down: {reciprocal(number)}",
        "20. Surd": f"An expression containing a root: {surd(number)}",
        "21. Fibonacci Number": f"A number in the Fibonacci sequence: {is_fibonacci(number)}",
        "22. Harmonic Number": f"A number that is the reciprocal of the arithmetic mean of the reciprocals of a set of numbers: {is_harmonic(number)}",
        "23. Perfect Number": f"A number equal to the sum of its proper divisors: {is_perfect(number)}",
        "24. Deficient/Abundant Number": f"A number for which the sum of its proper divisors is less than or greater than the number itself, respectively: {deficiency(number)}",
        "25. Lucky Number": f"A number generated by a specific operation on a set of integers: {is_lucky(number)}",
        "26. Mersenne Prime": f"A prime number that is one less than a power of two: {is_mersenne_prime(number)}",
        "27. Friendly Number": f"A number that shares a relationship with another number: {is_friendly(number)}",
        "28. Abundancy Index": f"The ratio between the sum of the proper divisors of a number and the number itself: {abundancy_index(number)}",
        "29. Aliquot Sum": f"The sum of the proper divisors of a number: {aliquot_sum(number)}",
        "30. Taxicab Number": f"A number that can be expressed as the sum of two positive cubes in two different ways: {is_taxicab_number(number)}",
        "31. Narcissistic Number": f"An n-digit number that is the sum of the nth powers of its digits: {is_narcissistic(number)}"
    }
    for desc, explanation in descriptions.items():
        print(desc)
        print(explanation)
        print()


def ordinal(num):
    if 10 <= num % 100 <= 20:
        suffix = 'th'
    else:
        suffix = {1: 'st', 2: 'nd', 3: 'rd'}.get(num % 10, 'th')
    return str(num) + suffix


def number_to_words(num):
    under_20 = ['Zero', 'One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten',
                'Eleven', 'Twelve', 'Thirteen', 'Fourteen', 'Fifteen', 'Sixteen', 'Seventeen', 'Eighteen', 'Nineteen']
    tens = ['', '', 'Twenty', 'Thirty', 'Forty', 'Fifty', 'Sixty', 'Seventy', 'Eighty', 'Ninety']
    above_100 = {100: 'Hundred', 1000: 'Thousand', 1000000: 'Million', 1000000000: 'Billion'}

    if num < 20:
        return under_20[int(num)]
    if num < 100:
        return tens[int(num) // 10] + ('' if num % 10 == 0 else ' ' + under_20[int(num) % 10])
    # find the appropriate pivot
    pivot = max([key for key in above_100.keys() if key <= num])
    return number_to_words(num // pivot) + ' ' + above_100[pivot] + ('' if num % pivot == 0 else ' ' + number_to_words(num % pivot))


def is_even(num):
    return num % 2 == 0


def is_odd(num):
    return not is_even(num)


def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True


def is_composite(num):
    return not is_prime(num) and num > 1


def is_positive(num):
    return num > 0


def is_negative(num):
    return num < 0


def is_zero(num):
    return num == 0


def is_integer(num):
    return num == int(num)


def is_decimal(num):
    return False


def is_fraction(num):
    return False


def is_rational(num):
    return num == int(num) or num == round(num)


def is_irrational(num):
    return not is_rational(num)


def is_real(num):
    return num == int(num) or num == round(num)


def is_imaginary(num):
    return not is_real(num) and num != 0


def is_complex(num):
    return is_real(num) or is_imaginary(num)


def absolute_value(num):
    return abs(num)


def scientific_notation(num):
    return "Not App"

"licable"


def factors(num):
    factors_list = []
    for i in range(1, num + 1):
        if num % i == 0:
            factors_list.append(i)
    return factors_list


def multiples(num):
    multiples_list = []
    for i in range(1, 11):
        multiples_list.append(num * i)
    return multiples_list


def divisibility(num):
    divisibility_list = []
    for i in range(2, num):
        if num % i == 0:
            divisibility_list.append(i)
    return divisibility_list


def powers(num):
    square = num ** 2
    cube = num ** 3
    higher_powers = [num ** i for i in range(4, 11)]  # Higher powers up to the 10th
    return f"Square: {square}, Cube: {cube}, Higher Powers: {higher_powers}"


def roots(num):
    root = num ** 0.5
    cube_root = num ** (1/3)
    higher_roots = [num ** (1/i) for i in range(4, 11)]  # Higher roots up to the 10th
    return f"Square Root: {root}, Cube Root: {cube_root}, Higher Roots: {higher_roots}"


def percentage(num):
    return "Not Applicable"


def decimal_places(num):
    return "Not Applicable"


def reciprocal(num):
    if num != 0:
        return 1 / num
    return "Undefined"


def surd(num):
    return f"√{num}"


def is_fibonacci(num):
    def is_perfect_square(x):
        return int(x ** 0.5) ** 2 == x
    return is_perfect_square(5 * num ** 2 + 4) or is_perfect_square(5 * num ** 2 - 4)


def is_harmonic(num):
    return num == 1


def is_perfect(num):
    if num < 2:
        return False
    factors_sum = sum(factors(num)[:-1])
    return factors_sum == num


def deficiency(num):
    factors_sum = sum(factors(num)[:-1])
    if factors_sum < num:
        return f"Deficient: {factors_sum}"
    elif factors_sum > num:
        return f"Abundant: {factors_sum}"
    else:
        return "Perfect"


def is_lucky(num):
    return True


def is_mersenne_prime(num):
    def is_prime(x):
        if x < 2:
            return False
        for i in range(2, int(x ** 0.5) + 1):
            if x % i == 0:
                return False
        return True

    p = 0
    while True:
        if is_prime(2 ** p - 1):
            if 2 ** p - 1 == num:
                return True
            elif 2 ** p - 1 > num:
                return False
        p += 1


def is_friendly(num):
    return False


def abundancy_index(num):
    factors_sum = sum(factors(num)[:-1])
    return factors_sum / num


def aliquot_sum(num):
    return sum(factors(num)[:-1])


def is_taxicab_number(num):
    def find_taxicab_number():
        n = 2
        while True:
            found = {}
            for a in range(1, n):
                for b in range(a, n):
                    result = a ** 3 + b ** 3
                    if result in found:
                        found[result].append((a, b))
                        if len(found[result]) == 2:
                            return result, found[result]
                    else:
                        found[result] = [(a, b)]
            n += 1

    result, pairs = find_taxicab_number()
    return num in pairs


def is_narcissistic(num):
    num_str = str(num)
    power = len(num_str)
    return num == sum(int(x) ** power for x in num_str)

if __name__ == "__main__":
    number = int(input("Enter a number: "))
    describe_number(number)
```
